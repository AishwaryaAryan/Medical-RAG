```python
from datasets import load_dataset

dataset = load_dataset(
    "medalpaca/medical_meadow_medqa",
    split="train[:1000]"
)

print(dataset)
```

```python
print(df.shape)
print(df.columns)
```

```python
df.to_csv(
    "../data/raw/medical_data.csv",
    index=False
)

print("Dataset Saved")
```

```python
documents = []

for _, row in df.iterrows():

    text = f"""
Question:
{row['input']}

Answer:
{row['output']}
"""

    documents.append(text)

print(documents[0])
```

```python
def chunk_text(text, chunk_size=300):

    chunks = []

    for i in range(0, len(text), chunk_size):

        chunks.append(
            text[i:i+chunk_size]
        )

    return chunks
```

```python
all_chunks = []

for doc in documents:

    all_chunks.extend(
        chunk_text(doc)
    )

print("Total Chunks:", len(all_chunks))
```

```python
import pickle

with open(
    "../data/processed/chunks.pkl",
    "wb"
) as f:

    pickle.dump(
        all_chunks,
        f
    )

print("Chunks Saved")
```
